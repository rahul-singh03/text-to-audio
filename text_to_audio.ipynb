```python
from gtts import gTTS
```

```python
from IPython.display import Audio
```

```python
def text_to_audio(text,language='en'):
  tts = gTTS(text = text, lang = language)
  tts.save('output.mp3')
```

```python
text = "जय श्री राम जय श्री राम जय श्री राम जय श्री राम जय श्री राम जय श्री राम जय श्री राम"
text_to_audio(text,'hi')
```

```python
Audio('output.mp3')
```

```python
from pydub import AudioSegment
```

```python
sound1 = AudioSegment.from_file('/content/drive/MyDrive/Hum katha sunate - Instrumental.mp3')
sound2 = AudioSegment.from_file('/content/output.mp3')
```

```python
combined = sound2.overlay(sound1)
```

```python
combined.export("combined.wav", format = 'wav')
```
